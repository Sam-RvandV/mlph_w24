```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def p_X(x):
    return 1/2 * x

def p_Y(y):
    return -1/2*y + 1
```

```python
# Generate samples that follow p_X
n = 10000
u = np.random.uniform(0, 1, size=n)
x_samples = 2 * np.sqrt(u)

# Plot
x = np.linspace(0, 2, 1000)
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# Plot pdf
ax[0].plot(x, p_X(x), color='red')
ax[0].set_xlabel('x')
ax[0].set_ylabel(r'$p_{X}(x)$')

# Plot histogram
ax[1].hist(x_samples, bins=30, color='skyblue', edgecolor='black')
ax[1].set_xlabel('x')
ax[1].set_ylabel(r'number of samples')

plt.show()
```

```python
# Transform samples
y_samples = 2 - np.sqrt(4 - x_samples**2)

# Plot
y = np.linspace(0, 2, 1000)
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# Plot pdf
ax[0].plot(x, p_Y(y), color='red')
ax[0].set_xlabel('y')
ax[0].set_ylabel(r'$p_{Y}(y)$')

# Plot histogram
ax[1].hist(y_samples, bins=30, color='skyblue', edgecolor='black')
ax[1].set_xlabel('y')
ax[1].set_ylabel(r'number of samples')

plt.show()
```
